```python
gold_data_cleaned = gold_data.drop(columns = ['Date'])
```

```python
correlation = gold_data_cleaned.corr()
```

Splitting the Features and Target


```python
X = gold_data.drop(['Date','GLD'], axis = 1)
Y = gold_data['GLD']
```

```python
print(X)
```

```python
print(Y)
```

Splitting into Training Data and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size= 0.2, random_state= 2)
```

Model Training :
Random  Forest Regressor

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
#training the model
regressor.fit(X_train,Y_train)
```

Model Evaluation

```python
# prediction on test data
test_data_prediction = regressor.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
# R squared error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared error : ", error_score)
```

Compare the actual values with predicted values in Plot

```python
Y_list = list(Y_test)
```

```python



plt.plot(Y_test.reset_index(drop=True), color='blue', label='Actual Value')
plt.plot(test_data_prediction, color='green', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of Values')
plt.ylabel('Gold Price')
plt.legend()
plt.show()
```
